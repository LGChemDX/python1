## **(1) 문자열 표현**

### **1) 기본 표현**

### **2) 여러 줄 문자열 표현**

- 문자열 앞뒤에 따옴표 세 개를 나열하거나, \\n을 사용해 여러 줄 표현이 가능합니다.
- print() 함수를 사용해야 의도한 내용이 제대로 표시됩니다.

```python
# 여러 줄 문자열 표시
s1 = "여러 줄의 문자열을 표시할 때는 \n을 사용하세요"
print(s1)
```

```python
s2 = """또 다른 방법은
큰 따옴표나 작은 따옴표를 세 개 사용하는 것입니다."""
print(s2)
```

### **3) 문자열 연산**

- 문자열 연산은 숫자 연산과 당연히 다릅니다.
- 문자열 연산에 대해 알아두면 이후에 편리하게 사용할 수 있습니다.

#### **① 문자열 + 문자열 = 문자열 연결**

```python
# 변수 선언
s1 = 'hello'
s2 = 'world'

# 문자열 연결
hw = s1 + ' ' + s2

# 확인
print(hw)
```

#### **② 문자열 * 정수 = 문자열 반복**

```python
# * 사용해 문자열 반복
print('*' * 17)
print('** ' + s1 + ' ' + s2 + ' **')
print('*' * 17)
```

```python
# 문자열에 0을 곱하면 문자열이 제거됨
s = 'hello'
print(s * 1)
print(s * 0)
print(s * 2)
```

## **(2) 표시 형식 지정**

### **1) format 메소드()**

- 문자열 일부분만 바꾸면서 해당 문자열을 재사용할 경우가 있습니다.
- 예를 들어 문자열에서 이름만 변경하며 사용하는 경우입니다.
- 이를 **문자열 포맷팅**이라고 하며, **문자열.format()** 형태로 사용합니다.

#### **① 기본 사용법**

- 단순 문자열에서 { }는 아무 의미가 없습니다.
- 하지만 format() 메소드를 만나면 { }는 큰 의미를 가집니다.

```python
# 단순 문자열
'이름:{}, 과목:{}, 점수:{}'
```

```python
# format 사용
'이름:{}, 과목:{}, 점수:{}'.format('홍길동', '국어', 90)
```

#### **② print() 함수 사용**

- format() 메소드는 대부분 print() 함수 안에서 사용합니다.

```python
# 변수 선언
name = '홍길동'
year = 2019
perf = 2345.6789

# 출력
print('이름:{}, 연도:{}, 실적:{}'.format(name, year, perf))
```

### **2) 자릿수와 정렬 방식 지정**

- { } 안에 변수 위치 값을 지정해야 하지만, 순서대로 사용될 것이면 생략하는 경우가 많습니다.
- { } 는 출력되는 값 순서 외에도 출력되는 형태, 즉 자리수 등을 제어할 수 있는 기능도 있습니다.

```python
# 변수 선언
name = '홍길동'
year = 2019
perf = 2345.6789

# 출력
print('이름:{0}, 연도:{1}, 실적:{2}'.format(name, year, perf))
print('이름:{0}, 연도:{1}, 실적:{2:.2f}'.format(name, year, perf))
print('이름:{}, 연도:{}, 실적:{:,.2f}'.format(name, year, perf))
```

### **3) f-String**

- 최근에는 다음과 같은 방법(f-String)을 많이 사용합니다.

```python
# 변수 선언
name = '홍길동'
age = 25
score = 2345.6789

# 확인
print(f'이름: {name}')
print(f'나이: {age}')
print(f'점수: {score}')
print(f'{age}살의 {name}의 점수는 {score}입니다.')
```

- 출력되는 형태를 나름으로 조절할 수 있습니다. (참고: 문자열:s, 정수:d, 실수:f)

```python
# 확인
print(f'{age}살의 {name}의 점수는 {score:,.2f}입니다.')
```

## **(3) 문자열 인덱싱과 슬라이싱**

### **1) 문자열 인덱싱**

- 대괄호([ ])안에 인덱스를 지정해 원하는 위치의 문자를 확인합니다.

```python
# 문자열 변수
s = '나의살던고향은꽃피는산골'

# 문자열 인덱싱
print(s[0])
print(s[3])
print(s[11])
print(s[-1])
print(s[-5])
```

### **2) 문자열 슬라이싱**

- 범위를 지정해 문자를 추출할 수 있습니다.
- 문자열[*m* : *n*] 형태의 구문으로 m부터 n-1번째 문자까지 추출합니다.


```python
# 문자열 변수
s = '나의살던고향은꽃피는산골'

# 문자열 슬라이싱
print(s[4:6])
print(s[7:])
print(s[-5:])
print(s[-5:len(s)])
print(s[:])
```

## **(4) 문자열 관련 주요 메소드**

- 문자열을 변환하는 메소드를 사용해도 그 결과가 변수에 반영되지는 않습니다.

```python
# 대문자로 변환
s = "I have a Dream."
print(s.upper())
print(s)
```


- 변환 결과를 변수에 반영하려면 메소드가 반환하는 값을 변수에 대입해야 합니다.

```python
# 대문자로 변환
s = "I have a Dream."
print(s.upper())

# 변수에 반영
s = s.upper()
print(s)
```

### **1) 대/소문자 변환**

- **capitalize()** 메소드: 문자열 첫 문자만 대문자, 나머지 모두 소문자로 변환합니다.
- **title()** 메소드: 단어 첫 문자만 대문자, 나머지는 모두 소문자로 변환합니다.
- **upper()** 메소드: 모든 문자를 대문자로 변환합니다.
- **lower()** 메소드: 모든 문자를 소문자로 변환합니다.

```python
# 대/소문자 변환
s = "I have a Dream."
print(s.capitalize())
print(s.title())
print(s.upper())
print(s.lower())
```

### **2) replace() 메소드**

- **replace(x, y)** 메소드: 문자열에 포함된 일부 문자열 x를 문자열 y로 변경합니다.

```python
# 문자열 변경
s = 'nowhere'
print(s.replace('now', 'now '))
```

- replace() 메소드를 사용하는 경우가 많으니 꼭 기억합니다.

```python
# 문자열 변수
num = '123,456'

# 콤마 제거 후 정수로 변환
num = num.replace(',', '')
num = int(num)

# 연산
print(num * 2)
```

### **3) split() 메소드**

- **split((x)** 메소드: x 문자를 구분자로 하여 문자열을 분리하여 리스트로 변환합니다.(공백이 기본 구분자)

```python
# 변수 선언
s = 'I have a Dream!'

# 분리
print(s.split())
```

```python
# 변수 선언
s = '2020-09-14'

# 분리
print(s.split('-'))
```

```python
# 변수 선언
s = '1/2/3/4/5/6/7/8/9'

# 분리
s_new = s.split('/')

# 확인
print(s_new)
```

```python
# 변수 선언
s = '2020-09-14'

# 분리
year = s.split('-')[0]
month = s.split('-')[1]
day = s.split('-')[2]

# 확인
print('오늘은', year, '년', month, '월', day, '일 입니다.')
```

### **4) join() 메소드**

- **x.join(s)** 메소드: x를 구분자로 하여 요소들을 연결한 문자열로 변환합니다.

```python
# 변수 선언
s = 'PYTHON'

# 구분자 삽입
s1 = '/'
print(s1.join(s))
```

```python
# 변수 선언
a = ['홍길동', '한사랑', '강우동', '박여인']

# 요소 연결
s = '/'.join(a)

# 확인
print(s)
```
