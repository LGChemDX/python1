## **(1) if 문**

- 가장 대표적인 제어문이 if 문입니다.
- 특정 조건에 맞는 경우에만 코드를 실행하고 싶을 때 사용압니다.

### **1) 단순 if 문**

- 조건이 참인 경우만 처리합니다.
- 조건이 거짓인 경우는 아무 처리도 하지 않습니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/if.png' width=200 align="left"/>

```python
# 변수 선언
my_score = 73

# 80보다 크거나 같으면 10 더하기
if my_score >= 80:
    my_score += 10

# 결과 출력
print(my_score)
```

```python
# 리스트 선언
score_list = []

# 리스트에 요소가 있으면 출력, 없으면 말고
if len(score_list) > 0:
    print(score_list)
```

```python
# 리스트 선언
score1 = [75, 80, 70, 95, 90]
score2 = []

# 요소가 있으면 True, 없으면 False
print(bool(score1), bool(score2))
```

```python
# 리스트 선언
score_list = [75, 80, 70, 95, 90]

# 리스트에 요소가 있으면 출력, 없으면 말고
if score_list:
    print(score_list)
```

### **2) if ~ else 문**

- 조건이 참인 경우와 참이 아닌 경우 각각 처리합니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/if_else.png' width=200 align="left"/>

```python
# 변수 선언
my_score = 73

# 80보다 크거나 같으면 Pass, 아니면 Fail
if my_score >= 80:
    my_result = 'Pass'
else:
    my_result = 'Fail'

# 결과 출력
print(my_result)
```

```python
# 리스트 선언
score_list = [75, 80, 70, 95, 90]

# 리스트에 요소가 있으면 출력, 없으면 '요소 없음' 출력
if score_list:
    print(score_list)
else:
    print('요소 없음')
```

- 처리할 구문이 한 줄이라면 다음과 같이 작성할 수 있습니다.

```python
# 리스트 선언
score_list = [75, 80, 70, 95, 90]

# 한 줄로 작성하기
if score_list: print(score_list)
else: print('요소 없음')
```

- 처리할 구문이 없거나 이후에 작성할 목적으로 **pass**를 사용할 수 있습니다.

```python
# 리스트 선언
score_list = [75, 80, 70, 95, 90]

# 우선 pass
if score_list:
    pass
else:
    print('요소 없음')
```

### **3) if ~ elif ~ else 문**

- 여러 조건 중에서 참인 경우 각각 처리, 그렇지 않은 경우 처리합니다.

<img src='https://raw.githubusercontent.com/jangrae/img/master/elif.png' width=200 align="left"/>

```python
# 변수 선언
my_score = 76

# 점수에 따른 학점 주기
if my_score >= 90:
    my_grade = 'A'
elif my_score >= 80:
    my_grade = 'B'
elif my_score >= 70:
    my_grade = 'C'
elif my_score >= 60:
    my_grade = 'E'
else:
    my_grade = 'F'

# 결과 출력
print(my_grade)
```

- 위 구문은 if ~ else 문을 사용해 다음과 같이 작성할 수 있습니다.
- 하지만 if ~ elif ~ else 문으로 작성하는 것이 더 가독성이 좋습니다.

```python
# 변수 선언
my_score = 76

# 점수에 따른 학점 주기
if my_score >= 90:
    my_grade = 'A'
else:
    if my_score >= 80:
        my_grade = 'B'
    else:
        if my_score >= 70:
            my_grade = 'C'
        else:
            if my_score >= 60:
                my_grade = 'E'
            else:
                my_grade = 'F'

# 결과 출력
print(my_grade)
```

## **(2) for 문**

- 첫번째 값부터 마지막 값까지 하나씩 가져가면서 처리합니다.

## **(3) while 문**

- 조건문이 True인 동안 while문 안의 문장이 반복해서 실행됩니다.
- 특정 상황에서 반복을 중지하게 while 문 안에서 조건을 제어합니다.

### **1) 일반 while 문**

- 문자열 길이를 반환하는 len() 함수가 리스트의 요소 개수를 반환합니다.

```python
# 리스트 선언
fruits = ['apple', 'banana', 'cherry']

# 요소 출력
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
```

```python
# 변수 선언
total = 0
i = 1

# 짝수만 더하기
while i <= 100:
    if i % 2 == 0:
        total = total + i
    i = i + 1
print(total)
```

### **2) 무한루프 다루기**

- break 문: 반복문 안에서 특정 조건이 되어 반복문을 빠져 나올 때 사용합니다.
- continue 문: 반복문을 중단시키지 않고 다음 반복으로 넘어갈 때 사용합니다.

```python
# 변수 선언
tot, i = 0, 0

# 100까지 짝수 합 구하기
while True:
    i += 1
    if i > 100:
        break

    if i % 2 == 0:
        tot += i
    else:
        continue

# 확인
print(tot)
```

## **(4) 리스트와 반복문**

- 반복문과 가장 많이 사용되는 자료형이 리스트형입니다.

### **1) 기본 반복문**

- 첫 번째 요소부터 마지막 요소까지를 대상으로 코드를 반복 처리합니다.

```python
# 리스트 선언
member = ['홍길동', '한사랑', '일지매', '박여인', '강우동']

# 루프 돌면서 이름 출력
for mem in member:
    print(mem)
```

### **2) enumerate() 함수 사용**

- 영어 단어 enumerate의 뜻은 '**낱낱이 세다**' 입니다.
- 문자열이나 컨테이너 자료형을 입력받아, 순번과 요소를 포함하는 오브젝트로 반환합니다.
- 순번과 요소를 반환하여 몇 번째, 어느 요소를 처리하고 있는지 확인 가능하게 합니다.

```python
list(enumerate(member))
```

```python
# 순번이 짝수인 경우만 '순번: 이름' 형태로 출력
for idx, mem in enumerate(member):
    if idx % 2 == 0:
        print(f'{idx}: {mem}')
```

```python
list(enumerate('나의살던고향'))
```

```python
# 문자열과 enumerate
for idx, val in enumerate('나의살던고향'):
    print(idx, val)
```

### **3) 반복문 확장**

- 이 방법을 사용하면 반복문 코드가 매우 간결해집니다.
- 처음에는 다소 낯설 수 있으나, 이것이 **Python 스러운 문법**이라 할 수 있습니다.

```python
# 리스트 선언
nums = [0, 1, 2, 3, 4]

# 빈 리스트 선언
squares = []

# 2를 곱한 결과를 squares 리스트에 추가
for x in nums:
    squares.append(x ** 2)

# 확인
print(squares)
```

```python
# 반복문 확장 문법 #1
squares = [x ** 2 for x in nums]

# 확인
print(squares)
```

```python
# 반복문 확장 문법 #2
# 반복문 + 조건문: 짝수인 값의 제곱을 구해서 even_squares 리스트에 넣기
even_squares = [x ** 2 for x in nums if x % 2 == 0]

# 확인
print(even_squares)
```
